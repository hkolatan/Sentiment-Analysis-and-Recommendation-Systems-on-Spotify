```python
def clean_dash_and_after(text):
    """
    This function takes a string and splits it by the dash, returning the first element of the split.
    """
    return text.split("-")[0]
```

```python
def clean_between_paranteses(text):
    """
    This function removes all text between parentheses from a string.
    """
    return re.sub(r'\([^)]*\)', '', text);
```

```python
def clean_missing_parantheses(text):
    return re.sub(r'\(.*$', '', text)
```

```python
df['name'] = df['name'].apply(lambda x: clean_dash_and_after(x))
```

```python
df["name"] = df["name"].apply(lambda x: clean_between_paranteses(x))
```

```python
df["name"] = df["name"].apply(lambda x: clean_missing_parantheses(x))
```

```python
df["name"] = df["name"].apply(lambda x: x.strip())
```

```python
df_filtered = df[~df['name'].str.isdigit()]
```

```python
df_filtered
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
sns.set_context("paper")

category_counts = df_filtered["category_name"].value_counts()
wedges, autotexts, _ = ax.pie(category_counts, colors=sns.color_palette("RdBu"), labels=['', ''], wedgeprops=dict(width=0.5), startangle=-40, autopct='%1.1f%%')

legend_labels = ["Active", "Relax"]
ax.legend(wedges, legend_labels, loc="upper center", bbox_to_anchor=(0.5, 0.55))

plt.setp(autotexts, size=10)
plt.title("Category Distribution", y=0.95, weight="bold")
plt.savefig("img/category_pie_chart.svg", dpi=300)
plt.show()
```
